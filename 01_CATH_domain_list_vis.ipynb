```python
# Classes

import matplotlib.pyplot as plt

alpha = 0
beta = 0
alpha_beta = 0
special = 0

cath_domain_data = '/Users/daxtraill/Desktop/honours_project/cath-domain-list.txt'

with open(cath_domain_data, 'r') as file:
    for line in file:
        parts = line.split()
        if len(parts) > 1:
            try:
                class_id = int(parts[1])
                if class_id == 1:
                    alpha += 1
                elif class_id == 2:
                    beta += 1
                elif class_id == 3:
                    alpha_beta += 1
                else:
                    special += 1
            except ValueError:
                continue


category_names = ['Mainly Alpha', 'Mainly Beta', 'Alpha-Beta', 'Special']
classes = [alpha, beta, alpha_beta, special]

fig, ax = plt.subplots()
ax.pie(classes, labels=category_names, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen', 'lightcoral'])
ax.axis('equal')

plt.title('Protein Classes')
plt.style.use('dark_background')
plt.show()
```
