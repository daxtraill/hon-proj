```python
# CATH cluster data S35, S60, S95, and S100

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

# ---------------------------------------------------------------------------------------------------
# Cluster data import

S35_data = '/volumes/dax-hd/project-data/cath-classification-data/cath-domain-list-S35.txt'
S60_data = '/volumes/dax-hd/project-data/cath-classification-data/cath-domain-list-S60.txt'
S95_data = '/volumes/dax-hd/project-data/cath-classification-data/cath-domain-list-S95.txt'
S100_data = '/volumes/dax-hd/project-data/cath-classification-data/cath-domain-list-S100.txt'
seq_data = '/volumes/dax-hd/project-data/cath-classification-data/cath-domain-list.txt'

# ---------------------------------------------------------------------------------------------------

data = {
    'S35': {'alpha': 0, 'beta': 0, 'alpha-beta': 0, 'special': 0},
    'S60': {'alpha': 0, 'beta': 0, 'alpha-beta': 0, 'special': 0},
    'S95': {'alpha': 0, 'beta': 0, 'alpha-beta': 0, 'special': 0},
    'S100': {'alpha': 0, 'beta': 0, 'alpha-beta': 0, 'special': 0},
    'Seq': {'alpha': 0, 'beta': 0, 'alpha-beta': 0, 'special': 0},
}

# ---------------------------------------------------------------------------------------------------

def process_file(file_path, cluster_key):
    # Add a print statement to debug file path issues
    print(f"Processing file: {file_path}")
    try:
        with open(file_path, 'r') as file:
            for line in file:
                # Debugging: Print lines to verify reading
                # print(line)
                parts = line.split()
                if len(parts) > 1:
                    try:
                        class_id = int(parts[1])
                        if class_id == 1:
                            data[cluster_key]['alpha'] += 1
                        elif class_id == 2:
                            data[cluster_key]['beta'] += 1
                        elif class_id == 3:
                            data[cluster_key]['alpha-beta'] += 1
                        else:
                            data[cluster_key]['special'] += 1
                    except ValueError:
                        continue
    except FileNotFoundError:
        print(f"File not found: {file_path}")

# Add checks or try/except blocks as necessary
process_file(S35_data, 'S35')
process_file(S60_data, 'S60')
process_file(S95_data, 'S95')
process_file(S100_data, 'S100')
process_file(seq_data, 'Seq')

# ---------------------------------------------------------------------------------------------------

categories = list(data[next(iter(data))].keys())  # Get categories from the first entry
groups = list(data.keys())

# Preparing the data
category_counts = [[data[group][category] for group in groups] for category in categories]

fig, ax = plt.subplots()
n_categories = len(categories)
n_groups = len(groups)
index = np.arange(n_categories)
bar_width = 0.15
opacity = 0.8

for i, group in enumerate(groups):
    plt.bar(index + i*bar_width, [data[group][category] for category in categories], bar_width,
            alpha=opacity, label=group)

plt.xlabel('Categories')
plt.ylabel('Counts')
plt.title('Counts by Class for Each Cluster Group')
plt.xticks(index + bar_width * (n_groups - 1) / 2, categories)
plt.legend(title="Cluster Groups", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.tight_layout()
plt.show()
```
