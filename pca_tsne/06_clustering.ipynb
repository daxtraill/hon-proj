```python
import pandas as pd
import matplotlib.pyplot as plt
from dpca import DensityPeakCluster

file = "/Volumes/dax-hd/project-data/images/tsne_topology_2/final_processed_data.csv"

data = pd.read_csv(file, header=0)

dpca = DensityPeakCluster(density_threshold=8, distance_threshold=7, anormal=False)

dpca.fit(data.iloc[:, [0, 10]])

dpca.plot("all", title="t-SNE - DPC")
fig, ax = plt.subplots(figsize=(20, 20))
categories = data['topology_description'].unique()
for category in categories:
    subset = data[data['topology_description'] == category]
    ax.scatter(subset.iloc[:, 0], subset.iloc[:, 1], label=category)
    ax.legend(title='Topology Description', bbox_to_anchor=(1.05, 1), loc='upper left')
    ax.set_xlabel('X Axis Label')
    ax.set_ylabel('Y Axis Label')
    plt.title('t-SNE - DPC Analysis with Categories')

print("t-SNE - DPC analysis completed.")
```
